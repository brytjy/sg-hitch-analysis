```python
import glob
from bs4 import BeautifulSoup
from tqdm import tqdm
```

```python
files = glob.glob("/home/bryan/Documents/jupyter/sg-hitch-analysis/SGHitch_06122019/*.html")
print("Total files: ", len(files))
for file in files[0:5]:
    print(file)
```

```python
for file in tqdm(files):
    f = open(file, "r")
    html = f.read()
    soup = BeautifulSoup(html)
    all_text = soup.findAll("div", {"class": "text"})
    f = open("sg_exam_chat.txt", "a")
    for text in all_text:
        f.write("%s\n" %text.text)
    f.close()
```

```python
f = open("/home/bryan/Documents/jupyter/sg-hitch-analysis/SGHitch_06122019/messages69.html", "r")
html = f.read()
soup = BeautifulSoup(html)
```

```python
all_messages = soup.findAll("div", {"class": "message default clearfix"})
```

```python
raw_datetime = []
raw_message = []

user_type = []
pick_up_location = []
drop_off_location = []
request_date = []
request_time = []
number_of_pax = []

for message in all_messages:
    bool_user_type = False
    bool_pick_up_location = False
    bool_drop_off_location = False
    bool_request_date = False
    bool_request_time = False
    bool_number_of_pax = False
    
    # Get date and time
    date_time = message.find("div", {"class": "pull_right date details"}).get("title")
    
    # Get text
    text = message.find("div", {"class": "text"})
    clean_text = str(text).replace('<div class="text">', "")
    clean_text = clean_text.replace('</div>', "")
    clean_text = clean_text.strip()
    clean_text = clean_text.lower()
    
    # Check if the text if really a request
    if "hitcher looking" in clean_text or "driver looking" in clean_text:
        # Store raw date and time
        raw_datetime.append(date_time)
        # Store raw text
        raw_message.append(clean_text)
        
        # PROCESS MESSAGE INTO STRUCTURED DATA
        text_tokenized = clean_text.split("<br/>")
        for token in text_tokenized:
            # User type
            if "hitcher looking" in token:
                user_type.append("hitcher")
            if "driver looking" in token:
                user_type.append("driver")
            # Pick up location
            if "pick:" in token or "pick up:" in token or "pickup:":
                loc = token.replace("pick:", "")
                loc = loc.replace("pick up:", "")
                loc = loc.replace("pickup:", "")
                loc = loc.strip()
                pick_up_location.append(loc)
            # Drop off location
            if "drop off:" in token or "dropoff:" in token:
                loc = token.replace("dropoff:", "")
                loc = loc.replace("drop off:", "")
                loc = loc.strip()
                drop_off_location.append(loc)
            # Date
            if "date:" in token:
                date = token.replace("date:", "")
                date = date.strip()
                if date == "today":
                    post_date = date_time.split(" ")[0].strip()
                    request_date.append(post_date)
                else:
                    request_date.append(date)
            # Time
            if "time:" in token:
                time = token.replace("time:", "")
                time = date.strip()
                if time == "now":
                    post_time = date_time.split(" ")[1][:5].strip()
                    request_time.append(post_time)
                else:
                    request_time.append(time)
            # Number of pax
            if "pax:" in token:
                pax = token.replace("pax:", "")
                pax = pax.strip()
                number_of_pax.append(pax)
```
